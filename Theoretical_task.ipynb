# Задание 3

***В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?***




### Решение

Для оценки результатов проведения события я бы использовал метрики : **ARPU, CR, ASL, Retention**.

Поскольку получение уникальных предметов напрямую зависит от количества проведенного в игре времени и получения уровней, то можно предположить, что в Plants & Gardens есть платные предметы, ускоряющие получение опыта в игре и получение очередного уровня.

Метрика **ARPU** позволит посмотреть, насколько увеличился заработок с каждого пользователя и насколько проведение таких событий (ограниченных по времени) подстегивает пользователей тратить деньги.

Метрика **Сonversion rate** позволит увидеть насколько данное событие смещает воронку конверсий от запуска игры и бесплатного получения ункальных предметов в сторону покупки. Таким образом можно предположить, можно ли такими событиями увеличить долю лояльных к игре (платящих) пользователей.

Метрика **ASL** показывает, насколько меняется количество времени проведенного пользователем в игре в зависимости от наличия или отсутсвия ограниченного по времени события. Если не платящий пользователь начинает проводить больше времени в игре, но не переходит в категорию платящих пользователей, то можно сделать вывод, что он лоялен к продукту и готов тратить на него время, но для его смещения в воронке конверсий в сторону покупки нужно искать другие способы воздействия.  Данная метрика вкупе с метрикой **Retention** позволяет определить таких пользователей в отдельный сегмент и проводить с ними дальнейшее A/B тестирование.

С помощью метрики **Retention** можно увидеть насколько сезонное событие увеличивает лояльность пользователей в дальнейшей перспективе. Если **Retention** возвращается к исходным значениям достаточно быстро, то необходимо пересматривать методы привлечения пользователей, повышения финансовых показателей. Метрика **Retention** может выступать в качестве индикатора проблем продукта. Если отток пользователей аномально сильный, необходимо тщательно изучить с чем это связано, возможно, в продукте встречаются баги и ошибки, которые отталкивают пользователей от дальнейшего использования продукта.

***Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?***

Если мы изменим механику, что при каждой неудачной попытке игрок будет откатываться на несколько уровней назад, я бы заменил метрику **Retention**, на обратную ей **Churn rate**. Поскольку мы вводим заведомо раздражающее событие, которое достаточно агрессивным образом склоняет пользователя к совершению покупки, я бы ожидал резкого ее возрастания, и подробно бы исследовал категории отваливающихся пользователей, среди них могут быть наиболее важные для нас лояльные платящие пользователи.
